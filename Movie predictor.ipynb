```python
def zyear(year):
    if year<1975:
        return 'very old'
    elif year<2005:
        return 'old'
    elif year<2018:
        return 'new'
    else:
        return 'latest'
movies['Movie age']=movies['release_year'].apply(zyear)
```

```python
def combine_col(row):
    try:
        return row['type']+' '+row['country']+' '+row['rating']+' '+row['listed_in']+' '+row['description']+' '+row['Movie age']
    except:
        print ("Error:",row)
    
```

```python
movies['combined_col']=movies.apply(combine_col,axis=1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cv=CountVectorizer()
count_matrix=cv.fit_transform(movies['combined_col'])
cosine_sim=cosine_similarity(count_matrix)
```

```python
movies['index']=movies.index
```

```python
def get_index_from_title(title):
    return movies[movies.title == title]['index'].values[0]
def get_title_from_index(index):
    return movies[movies.index == index]['title'].values[0]
```

```python
user_like_movie='10 Days in Sun City'
mov_index=get_index_from_title(user_like_movie)
similar_movies=list(enumerate(cosine_sim[mov_index]))
sorted_score=sorted(similar_movies, key=lambda x:x[1],reverse=True)
i=0
for movie in sorted_score:
    print(get_title_from_index(movie[0]))
    i=i+1
    if i>50:
        break
```
